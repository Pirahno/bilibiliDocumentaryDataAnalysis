### 得到记录片的所有信息

```python
def function(df):
    url = r'https://bangumi.bilibili.com/view/web_api/season?season_id={0}'.format(df['season_id'])
    print(url)
    r = requests.get(url)
    try:
        info = eval(r.text)['result']
        df['danmakus'] = info['stat']['danmakus']
        df['favorites'] = info['stat']['favorites']
        df['views'] = info['stat']['views']
        df['coins'] = info['stat']['coins']
        df['area'] = info['areas'][0]['name']
        df['style'] = info['style']
        df['episodes_aid'] = collections.OrderedDict()
        for episodes in info['episodes']:
            if episodes['section_type'] == 0:
                df['episodes_aid'][episodes['index_title']] = str(episodes['aid'])
        time.sleep(0.5)
    except:
        pass
    return df
```

## 评论时间分析

### 输入aid 得到该片的评论

```python
def get_time(item, delay = 0.5):
    info_list = []
    url = "http://api.bilibili.com/x/reply?type=1&oid=" + str(item) + "&pn=1&nohot=1&sort=0"
    print(url)
    r = requests.get(url)
    numtext = r.text
    json_text = json.loads(numtext)
    commentsNum = json_text["data"]["page"]["count"]
    if commentsNum%20 == 0:
        page = commentsNum // 20
    else:       
        page = commentsNum // 20 + 1
    for n in tqdm(range(page)):
        url = "https://api.bilibili.com/x/v2/reply?jsonp=jsonp&pn="+str(n+1)+"&type=1&oid="+str(item)+"&sort=1&nohot=1"
        req = requests.get(url)
        time.sleep(delay)
        text = req.text
        json_text_list = json.loads(text)
        for i in json_text_list["data"]["replies"]:
            info_list.append(i['ctime'])
    return info_list
#             if i['replies'] is not None:
#                 info_list.append(i['ctime'])
```

```python
def draw_video_replies(df, video_name):
    replies_time_list = collections.OrderedDict()
    try:    
        for each_name, each_episodes in eval(df[df['name'] == video_name]['episodes_aid'].iloc[0][11:]):
            print(each_name, each_episodes)
            replies_time_list[each_name] = get_time(each_episodes)
    except:
        raise NameError('Not found this name in dataframe')

    for name, each_time_list in replies_time_list.items():
    #     print(name, each_time_list)
        replies_time_list[name] = [time.strftime('%Y-%m-%d',time.localtime(i)) for i in each_time_list] 
    
    data_count =pd.DataFrame({'day':[]})#
    min_date = []
    for name, each_episode in replies_time_list.items():
        day_data =pd.DataFrame(each_episode, columns = ['day'])
        day_data[name] = '1'
        day_data_count = day_data.groupby(['day']).count().reset_index().head(40)
        day_data_count.columns = ['day', name]
        min_date.append(day_data_count['day'].min())
        data_count = pd.merge(data_count, day_data_count,how = 'outer',on=['day'])
    data_count['day'] = data_count['day'].astype('datetime64')
    data_count.set_index('day', inplace=True)# 设置日期格式为索引
# data_count = data_count.reset_index()
    ax = data_count.plot(figsize=(20,10), colormap=plt.cm.rainbow, linewidth=4,alpha=0.8)
    # 绘制每一集上传时间线
    for i in min_date:
        ax.vlines(i, 0, 2000, linestyle='--', color='blue')
    plt.xlabel('日期', size = 16)
    plt.ylabel('评论数量', size = 16, rotation = 0)
    plt.legend(fontsize  = 16)
    plt.title('')
    plt.show()
```

## 气泡图

2. x:弹幕数量
3. y:coins
4. 颜色深浅：（弹幕数量+投币数*10+收藏数*5）/播放数

```python
data_all_remove_zero_play = data_all[data_all['views']>100]
```

```python
data_all_remove_zero_play['rate'] = (data_all_remove_zero_play['coins']*10 + data_all_remove_zero_play['favorites']*5 + data_all_remove_zero_play['danmakus']*1)/(data_all_remove_zero_play['views'])
```

```python
from matplotlib import colors
norm = colors.Normalize(vmin= data_all_remove_zero_play['rate'].min(),vmax= data_all_remove_zero_play['rate'].max())
```

```python
ax = data_all_remove_zero_play.plot(figsize=(20, 12), kind ='scatter',x ='danmakus', y = 'coins',s = data_all_remove_zero_play['views']/5000 ,alpha = 0.5,c = data_all_remove_zero_play['rate'],norm = norm, colormap=plt.cm.Blues)
for (name, x_dir, y_dir) in zip(data_all_select['name'], data_all_select['danmakus'], data_all_select['coins']):
    plt.annotate(name, xy=(x_dir, y_dir), xytext=(x_dir-8000*len(name), y_dir),fontsize = 15)
    print(name)
plt.title('纪录片弹幕、投币')
plt.xlabel('弹幕数量', fontsize = 20)
plt.ylabel('投币数量', fontsize = 20, rotation = 0)
plt.show()
```
